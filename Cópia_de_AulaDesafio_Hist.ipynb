```python
# Importando as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#treinando importando as bibliotecas
import pandas as pd
import seaborn as sns 
```

```python
# Lendo dados tabulares

credit = pd.read_csv("/content/credit_card_clients.csv")
credit
```

```python
credit
```

```python
# Histograma

plt.figure(figsize=(10,10))

sns.histplot(credit, x="idade", hue="sexo", palette="light:m_r");

plt.legend(labels=["Masculino","Feminino"]);

# 1=Masculino, 2=Feminino
```

```python
#treinando histograma 
plt.figure(figsize=(10,10))

sns.histplot(data=credit,x='AGE',hue='SEX',kde=True,palette='light:m_r');
plt.legend(labels=['Masculino','Feminino'])
#funcao para montar um histograma 
```
