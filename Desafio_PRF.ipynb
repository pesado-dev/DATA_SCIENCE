```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd
import seaborn as sns
```

```python
# Carregar os dados
dados_2021 = pd.read_csv('datatran2021.csv', sep=';', encoding='latin1')
dados_2022 = pd.read_csv('datatran2022.csv', sep=';', encoding='latin1')
dados_2023 = pd.read_csv('datatran2023.csv', sep=';', encoding='latin1')
```

```python
# Concatenar os dados dos três anos em um único DataFrame
dados = pd.concat([dados_2021, dados_2022, dados_2023])
```

```python
dados
```

```python
# Estatísticas descritivas
print(dados.describe())
```

```python
valores_faltantes_por_coluna = dados.isnull().sum()
print("Valores faltantes por coluna:")
print(valores_faltantes_por_coluna)
```

```python
# Converter a coluna 'data_inversa' para datetime
dados['data_inversa'] = pd.to_datetime(dados['data_inversa'])

# Total de acidentes por ano
total_acidentes_por_ano = dados.groupby(dados['data_inversa'].dt.year)['id'].count()
print("Total de acidentes por ano:")
print(total_acidentes_por_ano)
print()
```

```python
# Total de acidentes por ano
total_acidentes_por_ano = dados.groupby(dados['data_inversa'].dt.year)['id'].count()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
total_acidentes_por_ano.plot(kind='bar', color='skyblue')
plt.title('Total de Acidentes por Ano')
plt.xlabel('Ano')
plt.ylabel('Total de Acidentes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
# Total de pessoas envolvidas por ano
total_pessoas_envolvidas_por_ano = dados.groupby('data_inversa')['pessoas'].sum()
print("Total de pessoas envolvidas por ano:")
print(total_pessoas_envolvidas_por_ano)
print()
```

```python
# Total de pessoas envolvidas por ano
total_pessoas_envolvidas_por_ano = dados.groupby(dados['data_inversa'].dt.year)['pessoas'].sum()

# Criar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(total_pessoas_envolvidas_por_ano.index, total_pessoas_envolvidas_por_ano.values, marker='o', linestyle='-')
plt.title('Total de Pessoas Envolvidas em Acidentes por Ano')
plt.xlabel('Ano')
plt.ylabel('Total de Pessoas Envolvidas')
plt.grid(True)
plt.xticks(total_pessoas_envolvidas_por_ano.index)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
# Total de acidentes por classificação do acidente
total_acidentes_por_classificacao = dados.groupby('classificacao_acidente')['id'].count()
print("Total de acidentes por classificação do acidente:")
print(total_acidentes_por_classificacao)
print()
```

```python
# Total de acidentes por classificação do acidente
total_acidentes_por_classificacao = dados.groupby('classificacao_acidente')['id'].count()

# Criar o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(total_acidentes_por_classificacao, labels=total_acidentes_por_classificacao.index, autopct='%1.1f%%', startangle=140)
plt.title('Total de Acidentes por Classificação do Acidente')
plt.axis('equal')
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
# Extrair o ano da coluna 'data_inversa'
dados['ano'] = dados['data_inversa'].dt.year

# Calcular os 5 principais tipos de acidentes em 2023
principais_tipos_acidentes_2023 = dados[dados['ano'] == 2023].groupby('tipo_acidente')['id'].count().nlargest(5)
print("5 principais tipos de acidentes em 2023:")
print(principais_tipos_acidentes_2023)
print()
```

```python
# Calcular os 5 principais tipos de acidentes em 2023
principais_tipos_acidentes_2023 = dados[dados['ano'] == 2023].groupby('tipo_acidente')['id'].count().nlargest(5)

# Criar o gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(principais_tipos_acidentes_2023.index, principais_tipos_acidentes_2023.values, s=100, c='blue', alpha=0.7)
plt.title('5 Principais Tipos de Acidentes em 2023')
plt.xlabel('Tipo de Acidente')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
# Qual rodovia com a maior média de acidentes ao longo dos 3 anos
media_acidentes_por_rodovia = dados.groupby('br')['id'].count().mean()
rodovia_maior_media_acidentes = dados.groupby('br')['id'].count().idxmax()
print("Rodovia com a maior média de acidentes ao longo dos 3 anos:", rodovia_maior_media_acidentes)
print()
```

```python
# Calcular a média de acidentes por rodovia
media_acidentes_por_rodovia = dados.groupby('br')['id'].count()

# Identificar a rodovia com a maior média de acidentes
rodovia_maior_media_acidentes = media_acidentes_por_rodovia.idxmax()

# Criar o gráfico de caixas
plt.figure(figsize=(10, 6))
sns.boxplot(x=media_acidentes_por_rodovia, color='skyblue')
plt.title('Distribuição da Média de Acidentes por Rodovia')
plt.xlabel('Média de Acidentes')
plt.grid(True, axis='x')

# Adicionar uma linha vertical para a rodovia com a maior média de acidentes
plt.axvline(x=media_acidentes_por_rodovia[rodovia_maior_media_acidentes], color='red', linestyle='--', linewidth=2, label='Rodovia com maior média')
plt.legend()

plt.show()

print("Rodovia com a maior média de acidentes ao longo dos 3 anos:", rodovia_maior_media_acidentes)

```

```python
# Qual rodovia com a maior média de óbitos ao longo dos 3 anos
media_obitos_por_rodovia = dados.groupby('br')['mortos'].sum().mean()
rodovia_maior_media_obitos = dados.groupby('br')['mortos'].sum().idxmax()
print("Rodovia com a maior média de óbitos ao longo dos 3 anos:", rodovia_maior_media_obitos)
print()
```

```python
import seaborn as sns

# Calcular a média de óbitos por rodovia
media_obitos_por_rodovia = dados.groupby('br')['mortos'].sum()

# Identificar a rodovia com a maior média de óbitos
rodovia_maior_media_obitos = media_obitos_por_rodovia.idxmax()

# Criar o gráfico de caixas
plt.figure(figsize=(10, 6))
sns.boxplot(x=media_obitos_por_rodovia, color='skyblue')
plt.title('Distribuição da Média de Óbitos por Rodovia')
plt.xlabel('Média de Óbitos')
plt.grid(True, axis='x')

# Adicionar uma linha vertical para a rodovia com a maior média de óbitos
plt.axvline(x=media_obitos_por_rodovia[rodovia_maior_media_obitos], color='red', linestyle='--', linewidth=2, label='Rodovia com maior média')
plt.legend()

plt.show()

print("Rodovia com a maior média de óbitos ao longo dos 3 anos:", rodovia_maior_media_obitos)
```

```python
dados['data_inversa'] = pd.to_datetime(dados['data_inversa'])
dados['dia_semana'] = dados['data_inversa'].dt.day_name()

# Adicionar a linha para converter a coluna 'horario' para o formato datetime
dados['hora'] = pd.to_datetime(dados['horario'], format='%H:%M:%S').dt.hour

# Calcular o dia da semana e o horário com o maior número de acidentes
dia_hora_maior_numero_acidentes = dados.groupby(['dia_semana', 'hora'])['id'].count().idxmax()
print("Dia da semana e horário com maior número de acidentes:", dia_hora_maior_numero_acidentes)
```

```python
# Calcular o número de acidentes por dia da semana e hora
acidentes_por_dia_hora = dados.groupby(['dia_semana', 'hora'])['id'].count().reset_index(name='num_acidentes')

# Encontrar o dia da semana e a hora com o maior número de acidentes
dia_hora_maior_numero_acidentes = acidentes_por_dia_hora.loc[acidentes_por_dia_hora['num_acidentes'].idxmax()]

# Criar o gráfico
plt.figure(figsize=(8, 6))
plt.bar(x=f"{dia_hora_maior_numero_acidentes['dia_semana']} às {dia_hora_maior_numero_acidentes['hora']}:00", height=dia_hora_maior_numero_acidentes['num_acidentes'], color='skyblue')
plt.title('Dia da Semana e Hora com Maior Número de Acidentes')
plt.xlabel('Dia da Semana e Hora')
plt.ylabel('Número de Acidentes')

plt.show()

print("Dia da semana e horário com maior número de acidentes:", dia_hora_maior_numero_acidentes['dia_semana'], "às", dia_hora_maior_numero_acidentes['hora'], "horas")
```

```python
# Calcular o número de acidentes por dia da semana e hora
acidentes_por_dia_hora = dados.groupby(['dia_semana', 'hora'])['id'].count().reset_index(name='num_acidentes')

# Encontrar o dia da semana com o maior número de acidentes
dia_maior_numero_acidentes = acidentes_por_dia_hora.groupby('dia_semana')['num_acidentes'].sum().idxmax()

# Filtrar os dados para o dia da semana com o maior número de acidentes
acidentes_dia_maior = acidentes_por_dia_hora[acidentes_por_dia_hora['dia_semana'] == dia_maior_numero_acidentes]

# Criar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(acidentes_dia_maior['hora'], acidentes_dia_maior['num_acidentes'], marker='o', color='skyblue', linestyle='-')
plt.title('Variação do Número de Acidentes ao Longo das Horas\n(Dia da Semana com Maior Número de Acidentes)')
plt.xlabel('Hora')
plt.ylabel('Número de Acidentes')
plt.xticks(range(24))
plt.grid(True)

plt.show()

print("Dia da semana com maior número de acidentes:", dia_maior_numero_acidentes)
```
