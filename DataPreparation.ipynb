```python
df_invest = pd.read_csv('/content/housing.csv')
```

```python
#usando a funo sort_values para ordernar os valor da coluna id
df_invest.sort_values(by= ['ID'])
```

```python
#com uma tabela booleana vai organizar por categoria e se for verdade cada linha vai retornar 0 ou 1
df_invest =  pd.get_dummies(df_invest, columns= ['backedLocation'])
```

```python
  #usando groupby para juntar colunas e somar especificas
  df_invest = df_invest.groupby(by=['ID']).agg({'age':'mean', 'backedLocation_BR': 'sum', 'backedLocation_GBK': 'sum', 'backedLocation_US': 'sum'})
```

```python
 #depois de ajustar valores e colunas, usando a funo merge para juntar a tabela em uma principal
 df =  df.merge(df_invest, how='right', on='ID')
```
